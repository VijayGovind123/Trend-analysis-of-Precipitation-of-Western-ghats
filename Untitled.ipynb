```python
def draw_screen_poly( lats, lons, m):
    x, y = m( lons, lats )
    xy = zip(x,y)
    poly = Polygon( list(xy),fill=None, edgecolor='black', alpha=0.5, linewidth=2, linestyle='--' )
    plt.gca().add_patch(poly)

lats = [ 8, 20, 20, 8 ]
lons = [ 72, 72, 77, 77 ]
```
